```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dimensionless time (0% to 100%)
t = np.linspace(0, 1, 100)

# Define a typical mass curve shape: use a Beta CDF
from scipy.stats import beta
dmc = beta.cdf(t, a=3, b=3)  # symmetric
# You can try (a=5, b=2) for early peaks, (a=2, b=5) for late peaks, etc.

# Scale to percentages
t_percent = t * 100
dmc_percent = dmc * 100

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(t_percent, dmc_percent, color='black', linewidth=2)

# Format
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_xlabel("% Dimensionless duration", fontsize=20)
ax.set_ylabel("% Dimensionless depth", fontsize=20)
ax.grid(True, linestyle=':', alpha=0.6)
ax.tick_params(axis='both', which='major', labelsize=15)

fig.savefig("../Figures/DMC_example.png")
plt.tight_layout()
plt.show()
```
