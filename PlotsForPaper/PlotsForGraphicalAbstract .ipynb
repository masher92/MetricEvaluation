```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gamma

# Set up the figure with dark background
plt.figure(figsize=(8, 6), facecolor='black')
ax = plt.subplot(111)
ax.set_facecolor('black')

# Generate time points
x = np.linspace(0, 10, 1000)

# Create the front-loaded event (using normal distribution)
front_loaded = norm.pdf(x, loc=2, scale=0.8) * 3

# Create the rear-loaded event (using gamma distribution)
rear_loaded = norm.pdf(x, loc=7, scale=0.9) * 1

# Combine the distributions
combined = front_loaded # + rear_loaded

# Plot the combined curve
# plt.plot(x, combined, color='white', linewidth=2)
plt.plot(x, front_loaded, color='royalblue', linewidth=2)
plt.plot(x, rear_loaded, color='orange', linewidth=2)

# Add labels with white text
plt.text(4.2, 1, 'Front-loaded\nevent', fontsize=20, color='royalblue', ha='center')
plt.text(7, 0.5, 'Back-loaded\nevent', fontsize=20, color='orange', ha='center')

# Set up axes labels and remove spines
plt.xlabel('Time', color='white', fontsize=12)
plt.ylabel('Rainfall intensity (mm/hr)', color='white', fontsize=12)
ax.set_title("Event asymmetry", fontsize=50, color='white')

# Remove top and right spines, and make bottom and left spines white
for spine in ax.spines.values():
    spine.set_visible(False)
    
# Add left and bottom spines back in white
ax.spines['left'].set_visible(True)
ax.spines['left'].set_color('white')
ax.spines['bottom'].set_visible(True)
ax.spines['bottom'].set_color('white')

# Remove tick marks but keep labels, and make them white
plt.tick_params(axis='both', colors='white', which='both', 
                bottom=False, top=False, left=False, right=False)


# Remove numerical tick labels
plt.xticks([])
plt.yticks([])

# Adjust layout
plt.tight_layout()


# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Setup figure
fig, axs = plt.subplots(1, 3, figsize=(18, 5), facecolor='black', sharey=True)
titles = ['     Event asymmetry', 'Peakiness', '   Concentration']
x1 = np.linspace(0, 10, 1000)
x2 = np.linspace(0, 10, 1000)
x3 = np.linspace(0, 10, 100)

# Panel 1: Asymmetry
front_loaded = norm.pdf(x1, loc=2, scale=0.8) * 8
rear_loaded = norm.pdf(x1, loc=7, scale=0.9) * 1
axs[0].plot(x1, front_loaded, color='royalblue', linewidth=2)
axs[0].plot(x1, rear_loaded, color='orange', linewidth=2)
axs[0].text(4.5, 3, 'Front-loaded\nevent', fontsize=20, color='royalblue', ha='center')
axs[0].text(7, 1, 'Back-loaded\nevent', fontsize=20, color='orange', ha='center')

# Panel 2: Peakiness
peaky_event = norm.pdf(x2, loc=5, scale=0.5) * 3
flat_event = norm.pdf(x2, loc=5, scale=1.5) * 3
axs[1].plot(x2, peaky_event, color='royalblue', linewidth=2)
axs[1].plot(x2, flat_event, color='orange', linewidth=2)
axs[1].text(2, 1.7, 'High peakiness', fontsize=20, color='royalblue', ha='center')
axs[1].text(9, 0.7, 'Low peakiness', fontsize=20, color='orange', ha='center')

# Panel 3: Temporal concentration
concentrated_event = np.array([0.5] * 100)
concentrated_event[45:55] = 4
np.random.seed(42)
scattered_event = np.random.choice([0, 1, 2, 3], size=len(x3), p=[0.6, 0.2, 0.15, 0.05])
axs[2].plot(x3, concentrated_event, color='royalblue', linewidth=2)
axs[2].plot(x3, scattered_event, color='orange', linewidth=2)
axs[2].text(2.2, 3.5, 'High\nconcentration', color='royalblue', fontsize=20, ha='center')
axs[2].text(9.5, 2.5, 'Low\nconcentration', color='orange', fontsize=20, ha='center')

# Styling all subplots
for ax, title in zip(axs, titles):
    ax.set_facecolor('black')
    ax.set_title(title, color='white', fontsize=40)
    ax.set_xlabel('Time', color='white', fontsize=12)
    ax.set_ylabel('Rainfall intensity (mm/hr)', color='white', fontsize=12)
    ax.spines['left'].set_visible(True)
    ax.spines['left'].set_color('white')
    ax.spines['bottom'].set_visible(True)
    ax.spines['bottom'].set_color('white')
    for spine in ['top', 'right']:
        ax.spines[spine].set_visible(False)
    ax.tick_params(axis='both', colors='white', bottom=False, left=False)
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
